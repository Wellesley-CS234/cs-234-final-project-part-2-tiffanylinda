```python
import re
import pandas as pd
import numpy as np

# Assuming your DataFrame is named 'df'
df_merge2['lyrics_cleaned'] = df_merge2['songs_lyrics'] 

def clean_lyric_ultimate_final(lyric_text):
    if not isinstance(lyric_text, str) or pd.isna(lyric_text):
        return lyric_text

    # --- Step 1: Perform the surgical split at the flexible prefix marker ---
    # This handles "Lyrics: Artist", "Artist Name:", "Lyrics Instrumentals", and language junk.
    # We look for a block that ends with a colon, "Lyrics", or "Instrumental".
    split_pattern = r'(lyrics\s*:\s*[\w\s]*)|([\w\s]*?[:])|(\w*lyrics\s*[:\s\-\(\)\w]*)|(\w*Instrumental\s*[:\s\-\(\)\w]*)'
    
    parts = re.split(split_pattern, lyric_text, maxsplit=1, flags=re.IGNORECASE | re.DOTALL)

    clean_lyric = lyric_text
    
    # Get the last non-empty string in the parts list, which is the clean content.
    if len(parts) > 1:
        for part in reversed(parts):
            if isinstance(part, str) and part.strip():
                clean_lyric = part.strip()
                break
        
    # --- Step 2: Suffix Cleaning ---
    # This removes promotional junk AND the "Instrumental : Etta James" style of tail.
    
    # NEW Suffix Pattern: Targets 'Instrumental', 'See Artist Live', 'Get tickets', etc.
    junk_suffix_pattern = r'(\s*Instrumental\s*[:\s\w]*|\s*See\s*[\w\s]+Live|\s*Get\s*tickets|\s*You\s*might\s*also\s*like|\s*Post\-\s*|^\s*$)[\w\W]*'
    match_suffix = re.search(junk_suffix_pattern, clean_lyric, re.IGNORECASE | re.DOTALL)
    
    if match_suffix:
        clean_lyric = clean_lyric[:match_suffix.start()].strip()

    # Final safe check for remaining '$' followed by digits (prices)
    clean_lyric = re.sub(r'\$\d+', '', clean_lyric).strip()
    
    return clean_lyric


# 1. Apply the ultimate cleaning function
df_merge2['lyrics_cleaned'] = df_merge2['songs_lyrics'].apply(clean_lyric_ultimate_final)

# 2. Drop the old corrupted column and rename the new one
df_merge2 = df_merge2.drop(columns=['songs_lyrics'])
df_merge2 = df_merge2.rename(columns={'lyrics_cleaned': 'lyrics'})

print("✅ Lyrics cleaning finalized. Proceeding to fix column order.")


# --- FINAL STEP: SAVE THE CLEANED BASELINE ---
df_merge2.to_csv('cleaned_lyrics2', index=False)
```

```python
import re
import pandas as pd
import numpy as np

# Assuming your DataFrame is named 'df'
df['lyrics_cleaned'] = df['lyrics'] 

def clean_lyric_by_replacement(lyric_text):
    if not isinstance(lyric_text, str) or pd.isna(lyric_text):
        return lyric_text

    # --- PART 1: Define Known Junk Patterns (Leading, Trailing, and Internal) ---

    junk_patterns = [
        # 1. The massive language/junk prefix (MUST be non-greedy and at start)
        # This targets the start of the string up to the *first* capitalized word/phrase that could be the start of a lyric.
        # This is the riskiest pattern, so we make it look for known junk markers.
        # Example: starts with non-word chars, followed by language names, up to "Lyrics" or a colon
        r'^[\W\s]*(?:Lyrics|Deutsch|Español|Türkçe|[\w\s]*?:)[\w\W]*?(?=[A-Z]{2,}\s|\n)',

        # 2. Specific "Artist Name :" or "Lyrics : Artist Name" patterns at the start of a line
        r'^\s*\w+\s*:\s*\w+[\s\w]*', # e.g., "Artist Name : Song Title"
        r'^\s*Lyrics\s*:\s*[\w\s]*', # e.g., "Lyrics : Etta James"

        # 3. Trailing junk (Promotional and Metadata)
        r'(\s*Instrumental\s*[:\s\w]*)+', # e.g., "Instrumental : Etta James"
        r'(\s*See\s*[\w\s]+Live|\s*Get\s*tickets|\s*You\s*might\s*also\s*like|\s*Post\-\s*)+[\w\W]*$', # Trailing promotions

        # 4. Prices
        r'\$\d+', 
        
        # 5. Simple, standalone markers that might have been left behind
        r'online\s+lyrics\s*:\s*',
        r'Lyrics\s*:\s*',
        r'Lyrics\s*Instrumental\s*',
        r'Lyrics\s*',
    ]

    clean_text = lyric_text

    # --- PART 2: Iterate and Replace ---
    for pattern in junk_patterns:
        # Use re.sub to replace the identified pattern with an empty string
        clean_text = re.sub(pattern, ' ', clean_text, flags=re.IGNORECASE | re.DOTALL).strip()
    
    # --- PART 3: Final Cleanup ---
    # Replace multiple spaces with a single space and strip whitespace
    clean_text = re.sub(r'\s+', ' ', clean_text).strip()
    
    return clean_text

# 1. Apply the new safer cleaning function
df['lyrics_cleaned'] = df['lyrics'].apply(clean_lyric_by_replacement)

# 2. Drop the old corrupted column and rename the new one
df = df.drop(columns=['lyrics'])
df = df.rename(columns={'lyrics_cleaned': 'lyrics'})
```

```python
import ast

# Ensure all list-like strings are converted to actual lists, 
# while safely handling non-strings (like NaN/None)
def safe_literal_eval(value):
    # Check for NaN (which is a float) or None
    if pd.isna(value) or value is None:
        return [] # Treat missing values as an empty list
    
    # If it is a string, evaluate it
    if isinstance(value, str):
        try:
            return ast.literal_eval(value)
        except (ValueError, SyntaxError):
            # If the string isn't a valid list format, just return it as is 
            # or return an empty list if you prefer to discard bad data
            print(f"Warning: Could not evaluate invalid string: {value}")
            return []
            
    # If it's already a list (or another non-string/non-null type), return it
    return value

# Apply the function to your DataFrame column
kaggle2['artists'] = kaggle2['artists'].apply(safe_literal_eval)
```

```python
import pandas as pd
import re

# --- 1. The Core Cleaning Function (Same as before) ---

def clean_lyrics(text_data: str) -> str:
    """
    Extracts and cleans song lyrics from a block of text, starting from 
    the "lyrics" marker (case-insensitive).
    """
    if not isinstance(text_data, str):
        # Handle potential non-string values (e.g., NaN)
        return ""

    # 1. Find the "lyrics" marker and everything after it
    match = re.search(r'lyrics\s*(.*)', text_data, flags=re.IGNORECASE | re.DOTALL)

    if not match:
        return ""

    raw_lyrics = match.group(1).strip()

    # 2. Clean the extracted lyrics by removing common trailing artifacts
    cleanup_pattern = re.compile(
        r'\s*(?:Edit|Report a problem|Thanks to .*\.)\s*$',
        re.IGNORECASE | re.DOTALL
    )
    cleaned_lyrics = cleanup_pattern.sub('', raw_lyrics)

    # 3. Final cleaning of whitespace and line breaks
    cleaned_lyrics = re.sub(r'[\r\n]+', '\n', cleaned_lyrics).strip()

    return cleaned_lyrics
```

```python
import pandas as pd
import re

def extract_lyrics_after_marker(text, marker="lyrics"):
    """
    Finds a marker (e.g., 'lyrics') in a string and returns all text 
    that appears immediately after it, stripping leading/trailing whitespace.
    
    Args:
        text (str): The input string from the Pandas cell.
        marker (str): The string used to define the starting point of the lyrics.
        
    Returns:
        str: The extracted lyrics, or the original text if the marker is not found.
    """
    if pd.isna(text) or not isinstance(text, str):
        return text  # Return non-string/NaN values as is

    # 1. Create a regex pattern that matches the marker case-insensitively
    #    and captures everything after it (.*)
    #    The 're.DOTALL' flag ensures that '.' also matches newline characters (\n).
    pattern = r'(.*?)(' + re.escape(marker) + r')(.*)'
    
    # 2. Search the text for the pattern
    match = re.search(pattern, text, re.IGNORECASE | re.DOTALL)
    
    if match:
        # Group 3 is the text captured after the marker
        extracted_text = match.group(3).strip()
        
        # Optional: Clean up common leading characters that might be left (e.g., ":")
        if extracted_text and extracted_text[0] in [':', '-', '>', '|', ' ']:
            extracted_text = extracted_text.lstrip(':->| ').strip()
            
        return extracted_text
    
    # If the marker is not found, return the original text
    return text.strip()
```
